```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-0.1,0.1)) 
```

```python
df = pd.read_csv('./sign_threshold.csv')
df['concurrent processing size'] = pd.to_numeric(df['concurrent processing size'], downcast="integer")
df['batch size'] = pd.to_numeric(df['batch size'], downcast="integer")
print(df.dtypes)
sns.set(rc = {'figure.figsize':(12,5)})
g = sns.lineplot(data=df,
             x='concurrent processing size',
             y='op/s',
             hue='batch size',
             palette='tab10',
             linewidth=0.5)
g.yaxis.set_major_formatter(formatter)
plt.savefig("./fig-res-throughput.pdf")
```

```python
sns.set_theme(style="whitegrid")
data = pd.read_csv('./dili2_batch.csv')
sns.set(rc = {'figure.figsize':(5,5)})
g = sns.lineplot(data=data,
             x='batch size',
             y='op/s',
             hue='procedure',
             linewidth=1.5)
g.yaxis.set_major_formatter(formatter)
plt.savefig("./fig-res-batch.pdf")
```

```python
sns.set_theme(style="whitegrid")
data = pd.read_csv('./dili2_stream.csv')
sns.set(rc = {'figure.figsize':(5,5)})
g = sns.lineplot(data=data,
             x='number of streams',
             y='op/s',
             hue='procedure',
             linewidth=1.5)
g.yaxis.set_major_formatter(formatter)
plt.savefig("./fig-res-stream.pdf")
```
